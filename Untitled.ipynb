```python
from imgaug import augmenters as iaa
from PIL import Image
import cv2
```

```python
img = cv2.imread('./output/test2/imgs/img_100.jpg')
seq = iaa.Sequential([
    iaa.BlendAlphaMask(
        iaa.InvertMaskGen(0.5, iaa.VerticalLinearGradientMaskGen()),
        iaa.Sequential([
#             iaa.MotionBlur(k=(3, 7)),
            iaa.AverageBlur(k=(3, 7)),
            iaa.GammaContrast(gamma=(.1, 3)),
                       ])
    ),
#     iaa.ElasticTransformation(alpha=1500, sigma=200, order=1),
#     iaa.ElasticTransformation(alpha=1500, sigma=200, order=1),
#     iaa.ElasticTransformation(alpha=1000, sigma=100, order=1),
])
aug_img = seq(image=img)
Image.fromarray(aug_img)
```
