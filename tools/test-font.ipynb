```python
from PIL import Image, ImageFont, ImageDraw
import os
```

```python
def test_font(font_dir):
  image_width = 700
  image_height = 5900
  image = Image.new('RGB', (image_width, image_height,), '#fff')
  draw = ImageDraw.Draw(image)

  fonts = [f for f in os.listdir(font_dir) if f.endswith('.ttf')]
  fonts.sort()
  curr_y = 10
  for font_name in fonts:
    font = ImageFont.truetype(f'{font_dir}/{font_name}', 20)
    text = font_name + ' 1234567890 เก้าสิบห้าร้อยสามสิบณฐชารย์'
    draw.text((10, curr_y), text, font=font, fill='#000')
    print(font_name)
    x0, y0, x1, y1 = draw.textbbox((10, curr_y), text, font=font)
    text_height = font.getsize(font_name)[1] + 5
    draw.rectangle((x0, y0, x1, y1), outline='#f00')
    curr_y += text_height
  return image
```

```python
test_font('../fonts/handwriting')
```

```python
test_font('../fonts/formal')
```
